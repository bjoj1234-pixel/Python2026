```python
from bs4 import BeautifulSoup
#파일 읽기
with open('greenData/member_data.xml','r',encoding='utf-8-sig')as f:
    soup = BeautifulSoup(f,'xml')
#데이터추출(select 모두출력: 내가 지정한 member(반드시 태그여야함)안의 모든 것)
members = soup.select('member')

for m in members:
    name = m.select_one('name').text
    age = m.select_one('age').text
    city = m.select_one('city').text
    print(f"이름:{name}, 나이:{age}, city:{city}")
    
```

```python
#pandas 라이브러리란:  테이블형 데이터를 다룰수 있는 다양한 기능을 가진 라이브러리이다.
#raw data: 아직 데이터 분석을 위해 정제되지않은 기본데이터를 의미함
#pandas는 반드시 import해서 사용한다.
# 데이터를 다루는 방식 2가지
    #시리즈(series)는 1차원(행하나 또는 열하나)
    #데이터프레임(Dataframe)은 2차원(테이블형) 데이터이다.

import pandas as pd #라이브러리 임포트(대부분 pd로 사용)
#pd.판다스 함수 형태로 호출하여 사용한다.
#index는 행의 레이블(제목)을 의미한다.
#index를 지정하지 않으면 0부터 시작하는 인덱스 자동생성
#지정할 경우는 지정된 index로 사용
#seriesdata = pd.Series([70,80,90]) => 미지정의 경우 index는 0,1,2
seriesdata = pd.Series([70,80,90], index=['국어','영어','수학'])
seriesdata #print() 생략해도 똑같이 나옴
seriesdata.index #Index(['국어', '영어', '수학'], dtype='object') => object는 문자(string)라는뜻
seriesdata.index = ["미술","과학","사회"]
seriesdata #미술    70, 과학    80,사회    90

#Series 삭제 가능 : del seriesdata['미술']
del seriesdata['미술']
seriesdata #과학    80,사회    90
```

```python
import pandas as pd

data = {
    '대여소':['강남','강남','홍대','홍대','잠실','잠실'],
    '이용자수':[120,150,80,90,200,100],
    '대여시간':[30,45,20,25,60,55],
    '이용일자':['2024-01-01','2024-01-02','2024-01-01','2024-01-02','2024-01-01','2024-01-02']
}
df = pd.DataFrame(data)
```

```python
#Data자료를 이용해 이용자수가 100명 이상인 데이터만 출력하시오
#조건필터
#힌트: df[df['age'] > 20]

df[df['이용자수'] >= 100]
```

```python
# 문제: 대여시간 기준으로 오름차순 정렬하여 출력하시오.
# 힌트: df.sort_values(by='price') =>디폴트값: 오름차순
df.sort_values(by='대여시간')
#내림차순 정렬: df.sort_values(by='price', ascending=False) => 값을 넣으면 내림차순
df.sort_values(by='대여시간', ascending=False)
```

```python
#문제 : 총 이용량 컬럼을 이용자수 * 대여시간으로 계산하여 추가하시오.

df['총 이용량'] = df['이용자수'] * df['대여시간']
#실제 데이터에 추가한게 아니라, 출력만 시킨거임
df
```

```python
# 문제 : Data자료를 이용해 대여소별 총 이용자수 구하여 출력하시오.
# 사용함수 : groupby()

result = df.groupby('대여소')['이용자수'].sum()

#=> 이 대여소별 이용자수를 내림차순하여 출력하시오.
result.sort_values(ascending=False)
```

```python
#문제: Data자료를 이용해 이용자수가 100 이상인 데이터만 추출하여 대여소별 평균 대여시간 구하여 출력하시오.
hundred = df[df['이용자수'] >= 100]
hundred.groupby('대여소')['대여시간'].mean()
```

```python
#문제: Data자료를 이용해 대여소별 "총이용량" 합계구하기
df['총 이용량'] = df['이용자수'] * df['대여시간']
print("대여소별 총 이용량 합계")
df.groupby('대여소')['총 이용량'].sum()

#문제: 가장 바쁜 날짜 찾기. => 이용일자별 이용자수 합계가 가장많은 날짜 출력
result02 = df.groupby('이용일자')['이용자수'].sum().sort_values(ascending=False)
print("가장 바쁜 날짜")
print(result02)
print(f"가장 바쁜 날짜: {result02.index[0]}")

#문제: 이용자 수 평균이 가장 높은 대여소찾기.
result03 = df.groupby('대여소')['이용자수'].mean().sort_values(ascending=False)
print("대여소별 평균 이용자수")
print(result03)
print(f"평균 이용자수 가장 높은 대여소: {result03.index[0]}")
```
