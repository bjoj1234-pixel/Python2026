--기본문법
open('파일명', '모드', encoding='utf-8-sig')
r: 읽기모드
w: 쓰기모드(새로생성, 기존내용삭제)
a: 추가모드(기존 내용 뒤에 이어 붙이기)

문제] score_data.csv 파일의 데이터중 국어점수가 85점 이상인 데이터만 추출하여 출력하세요
## DictReader => csv 파일을 읽기 좋게 변환하는데 첫 줄을 제목(=필드명)으로 인식하도록 변경해준다.

```python
import csv

#1. 파일 읽기
with open('greenData/score_data.csv','r',encoding='utf-8-sig') as f:
    #2. CSV의 첫줄을 필드명으로 인식하도록 DictReader 이용한다.
    reader = csv.DictReader(f)
    # reader = csv.reader(f) => print(r[1]) => '국어'...

    print('국어 점수가 85점 이상인 데이터:')

    #3. 한 줄씩 확인하면서 if조건검사
    for r in reader:
        # 파일에서 읽어온 모든 숫자는 문자열이므로 반드시 int()로 숫자형으로 자료를 변환한다.
        if int(r['국어']) >= 85:
            print(r)
```
