```python
#pandas 라이브러리는 사용시 반드시 import
import pandas as pd

#출력옵션 설정
pd.set_option('display.max_columns',None) #모든컬럼 출력
pd.set_option('display.width',1000) #출력너비 지정
pd.set_option('display.unicode.east_asian_width',True) #한글정렬 맞추기

#csv파일은 공공데이터포털에서 다운로드 받은자료임. => encoding='cp949'
#bike02.csv를 pandas를 이용해서 읽기
df = pd.read_csv('greenData/pm10.csv', encoding='cp949')

#'평균'인 행을 제외하고 출력
disc = df[df['구분'] != '평균']
#다중 조건 줄때 and : &, or : **, not : ~
condition = (disc['미세먼지(PM10)'] >= 25) & (disc['초미세먼지(PM25)'] >= 15)
cond02 = disc[condition]
result = cond02[['구분','미세먼지(PM10)','초미세먼지(PM25)']]
print("--- 미세먼지 및 초미세먼지 주의 지역 ---")
print(result)



```

```python
#csv에 입력된 날짜는 문자열 데이터 타입이다
#날짜형식으로 형변환하는 함수 : datetime()
import pandas as pd
#from datetime import datetime

#출력옵션 설정
pd.set_option('display.max_columns',None) #모든컬럼 출력
pd.set_option('display.width',1000) #출력너비 지정
pd.set_option('display.unicode.east_asian_width',True) #한글정렬 맞추기

#csv파일은 공공데이터포털에서 다운로드 받은자료임. => encoding='cp949'
#bike02.csv를 pandas를 이용해서 읽기
df = pd.read_csv('greenData/accident.csv', encoding='cp949')

# 문자열 → 날짜로 변환
# datetime import 없이 사용하려면 앞에 'to_'를 붙이면 된다.
df['사고일자'] = pd.to_datetime(df['사고일자'])

# 사고일자의 월을 출력
# dt.month:월, dt.year:년, dt.day:일, dt.hour:시, dt.second:초, dt.minute:분
# dt.dayofweek: 요일을 숫자로 반환(0:월요일 ~ 6:일요일)
# dt.day_name(): 요일 이름이 문자열 ('Monday')
# dt.quarter: 분기(1분기 ~ 4분기)
df['사고월'] = df['사고일자'].dt.month


# 월별사고 건수를 집계하여 어느월에 사고가 가장많이 발생했는지 확인하세요.
# groupby().count() => 데이터가 비어있지 않은 값의 개수를 센다.
# groupby().size() => 그룹별 전체 행의 개수를 센다.(비어있는 값도 개수센다)

# print("--- 월별 사고 발생 건수 ---")
# monthCount = df.groupby('사고월').size()
# monthCount

print("--- 사고가 가장 많은 월 ---")
# 정답1
# monthCount02 = df.groupby('사고월').size().sort_values(ascending=False)
# print(f"{monthCount02.index[0]}월")

# 정답2
monthCount02 = df.groupby('사고월').size()
maxMonth = monthCount02.idxmax() #인덱스 매치 함수 : 가장 큰 값이 들어있는 위치(인덱스 이름)에 접근하여 가져옴
print(f"{maxMonth}월")

# 사망자수와 부상자수를 더하여 '사상자수'라는 파생변수를 생성하세요.
df['사상자수'] = df['사망'] + df['부상']

# 노선명별로 그룹화하여 '사상자수'의 총합을 구하고,
# 사상자가 가장 많이 발생한 상위 5개 노선을 내림차순으로 출력하세요.

result = df.groupby('노선명')['사상자수'].sum().sort_values(ascending=False).head(5)
print("-- 사상자수 기준 상위 5개 위험 노선 --")
result

# 사고원인이 '졸음'이면서 사망자가 1명이상 발생한 치명적인 사고는 총 몇건인지 구하시오
result02 = len(df[(df['원인'] == '졸음') & (df['사망'] >= 1)]) #group을 안묶었기때문에 count() 쓰지않고 len을 씀
print("-- 졸음 운전중 사망자가 발생한 치명적 사고 건수 --")
result02
```
